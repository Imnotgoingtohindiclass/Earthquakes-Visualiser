```python
import requests
import pandas as pd

url = "https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/significant_month.csv"
df = pd.read_csv(url)
print(df.head())
```

```python
import matplotlib.pyplot as plt

# Scatter plot of earthquake locations
plt.figure(figsize=(10, 6))
plt.scatter(df['longitude'], df['latitude'], c=df['mag'], cmap='Reds', alpha=0.6)
plt.colorbar(label='Magnitude')
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Recent Significant Earthquakes")
plt.show()
```

```python
import plotly.express as px

fig = px.scatter_geo(df,
                    lat='latitude',
                    lon='longitude',
                    color='mag',
                    size='mag',
                    hover_name='place',
                    title="Interactive Earthquake Map",
                    projection="natural earth")

fig.update_layout(showlegend=True)
fig.write_html("earthquake_map.html")
```
